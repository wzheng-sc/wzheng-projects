```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.stats as sp
```

### Minimum Required Sample Size

```python
def min_sample_size(alpha, beta, mde, sigma):
    ppf = sp.norm(0, 1).ppf
    z_alpha = ppf(1-alpha/2)
    z_beta = ppf(1 - beta)
    n = int((2*sigma*(z_alpha + z_beta)/mde)**2)

    print(f"The minimum sample size is: {n}.")

mde = 1
sigma = 6.68
alpha = 0.05
beta = 0.2

min_sample_size(alpha, beta, mde, sigma)
```

### Naive, Bonferroni Correction and Group Sequential
- Spotify Engineering Blog: [Choosing a Sequential Testing Framework â€” Comparisons and Discussions](https://engineering.atspotify.com/2023/03/choosing-sequential-testing-framework-comparisons-and-discussions/)
- Uber's paper: [Safely and Quickly Deploying New Features with a Staged Rollout Framework Using Sequential Test and Adaptive Experimental Design](https://arxiv.org/pdf/1905.10493)

```python
def generate_data(n_ctrl, n_trt, mu_ctrl, sigma_ctrl, mu_trt, sigma_trt):
    X_ctrl = np.random.normal(mu_ctrl, sigma_ctrl, n_ctrl)
    X_trt = np.random.normal(mu_trt, sigma_trt, n_trt)
    return X_ctrl, X_trt

def compute_means_stdevs(sample_ctrl, sample_trt):
    n, k = np.shape(sample_ctrl)
    ns = np.arange(1, n + 1).reshape(-1, 1)
    means_ctrl = np.cumsum(sample_ctrl, axis=0) / ns
    means_trt = np.cumsum(sample_trt, axis=0) / ns
    stdevs_ctrl = np.sqrt(np.cumsum((sample_ctrl - means_ctrl) ** 2, axis=0) / ns)
    stdevs_trt = np.sqrt(np.cumsum((sample_trt - means_trt) ** 2, axis=0) / ns)
    return means_ctrl, means_trt, stdevs_ctrl, stdevs_trt

def update_intervals(df_intervals, i, means_ctrl, means_trt, stdevs_ctrl, stdevs_trt, alpha, mu, n_peek):
    ppf = sp.norm.ppf
    mean_diff = means_trt[i - 1, :] - means_ctrl[i - 1, :]
    pooled_std = np.sqrt(stdevs_ctrl[i - 1, :]**2 + stdevs_trt[i - 1, :]**2)
    df_intervals[f"mean_diff{i}"] = mean_diff
    df_intervals[f"width{i}"] = ppf(1 - alpha / 2) * pooled_std / np.sqrt(i)
    df_intervals[f"lowerb{i}"] = mean_diff - df_intervals[f"width{i}"]
    df_intervals[f"upperb{i}"] = mean_diff + df_intervals[f"width{i}"]
    df_intervals[f"coverage{i}"] = (df_intervals[f"lowerb{i}"] <= mu) & \
                                   (df_intervals[f"upperb{i}"] >= mu)
    df_intervals["rejected_0"] = (df_intervals[f"lowerb{i}"] >= 0) | \
                                 (df_intervals[f"upperb{i}"] <= 0) | \
                                 (df_intervals["rejected_0"])
    df_intervals[f"power{i}"] = df_intervals["rejected_0"]
    df_intervals["rejected_1"] = df_intervals["rejected_1"] | ~df_intervals[f"coverage{i}"]
    df_intervals[f"falsepos{i}"] = df_intervals["rejected_1"]
    df_intervals["length"] = np.minimum(df_intervals["length"], i) * df_intervals["rejected_0"] \
                             + max(n_peek) * (1 - df_intervals["rejected_0"])
    

def peek_intervals(sample_ctrl, sample_trt, n_peek, choose_alpha, alpha=0.05, mu=1):
    means_ctrl, means_trt, stdevs_ctrl, stdevs_trt = compute_means_stdevs(sample_ctrl, sample_trt)
    n, k = np.shape(sample_ctrl)

    df_intervals = pd.DataFrame({"k": range(k)})
    df_intervals["rejected_0"] = False
    df_intervals["rejected_1"] = False
    df_intervals["length"] = max(n_peek)

    for i in n_peek:
        current_alpha = choose_alpha(i, n, n_peek)
        update_intervals(df_intervals, i, means_ctrl, means_trt, stdevs_ctrl, stdevs_trt, 
                         current_alpha, mu, n_peek)

    return df_intervals


def plot_peeking(df, sample_ctrl, sample_trt, k=3):
    N, j = np.shape(sample_ctrl)
    ns = np.reshape(np.arange(1, N+1), (-1, 1))
    means_ctrl = np.cumsum(sample_ctrl, axis=0) / ns
    means_trt = np.cumsum(sample_trt, axis=0) / ns
    mean_diff = means_trt - means_ctrl
    n_peek = np.arange(200, N+1, 200)

    fig, ax = plt.subplots()
    mpl.rcParams['figure.figsize'] = (15, 4)
    for n in n_peek:
        color = "C1" if df[f"width{n}"][k] > df[f"mean_diff{n}"][k] else "C2"
        ax.errorbar(x=[n-1], y=df[f"mean_diff{n}"][k], yerr=df[f"width{n}"][k], c=color, lw=0, elinewidth=2, capsize=6, marker="o", markersize=6)
    sns.lineplot(x=range(20, N), y=mean_diff[:,k][20:], lw=3, ax=ax)
    ax.axhline(y=0, lw=1, c="Red", linestyle = "dotted")
    ax.set(xlim=(20, N+10), xlabel="Step", title="Average Treatment Effect and Confidence Intervals")


def check_performance(df):
    df_perf = df[list(df.filter(regex="coverage|power|falsepos"))].copy()
    df_perf["idx"] = 1
    df_perf = df_perf.groupby("idx", as_index=False).mean()
    df_perf = pd.wide_to_long(df_perf, stubnames=["coverage", "power", "falsepos"], i="idx", j="n") \
                .reset_index().drop(columns="idx")
    return df_perf


def plot_power(df, method, alpha=0.05, beta=0.8):
    df_perf = check_performance(df)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 3))

    #power
    sns.lineplot(df_perf, x="n", y="power", c="C1", ax=ax1)
    ax1.axhline(y=beta, ls="--", lw=2, c="C1")
    ax1.set(title=f"Power - {method}", xlabel = 'Step', ylabel="Power", ylim=(0.00, 1.05))

    #FPR
    sns.lineplot(df_perf, x="n", y="falsepos", c="C2", ax=ax2)
    ax2.axhline(y=alpha, ls="--", lw=2, c="C2")
    ax2.set(title=f"False Positive - {method}", xlabel = 'Step', ylabel="False Positive Rate", ylim=(0.0, 0.5))
```

### Insert Parameters

```python
#parameters
mu_ctrl = 0.0
mu_trt = 1.0
sigma_ctrl = sigma_trt = sigma
ate = mu_ctrl - mu_trt
k = 1000
n = 1400

#generate data using the `generate_data` function
sample_ctrl_a = np.zeros((n, k))
sample_ctrl_b = np.zeros((n, k))
sample_trt = np.zeros((n, k))

for i in range(k):
    sample_ctrl_a[:, i], sample_ctrl_b[:, i] = generate_data(n, n, mu_ctrl, sigma_ctrl, mu_ctrl, sigma_ctrl)
    _, sample_trt[:, i] = generate_data(n, n, mu_trt, sigma_trt, mu_trt, sigma_trt)

n_peek = np.arange(200, n + 1, 200)
print(f"We will peek at {n_peek}.")
```

### Naive Method

```python
def naive_alpha(i, n, n_peek, alpha=0.05):
    return alpha

df_naive = peek_intervals(sample_ctrl_a, sample_trt, n_peek, naive_alpha)
plot_peeking(df_naive, sample_ctrl_a, sample_trt)
plot_power(df_naive, 'Naive')
print(f"Average length: {df_naive.length.mean():.0f}")
```

### Bonferroni Correction

```python
def bonferroni_alpha(i, n, n_peek, alpha=0.05):
    n_p = len(n_peek)
    return alpha/n_p

df_bonferroni = peek_intervals(sample_ctrl_a, sample_trt, n_peek, bonferroni_alpha)
plot_power(df_bonferroni, "Bonferroni Correction")
print(f"Average length: {df_bonferroni.length.mean():.0f}")
```

### Group Sequential Test - Oâ€™Brien & Fleming Approximation

```python
def gst_obrien_fleming_alpha(i, n, n_peek, alpha=0.05, rho=0.5):
    cdf = sp.norm(0, 1).cdf
    ppf = sp.norm(0, 1).ppf
    t = i / n
    return 4 - 4 * cdf(ppf(1 - alpha/4) / t**(rho/2))

df_gst = peek_intervals(sample_ctrl_a, sample_trt, n_peek, gst_obrien_fleming_alpha)
plot_power(df_gst, "GST - O'Brien-Fleming")
print(f"Average length: {df_gst.length.mean():.0f}")
```

### Compare Bonferroni-corrected Alpha vs Oâ€™Brien & Fleming Alpha

```python
def compare_alpha(n, n_peek, alpha=0.05, rho=0.5):
    obrien_fleming_alpha = []
    for i in n_peek:
        of_alpha = gst_obrien_fleming_alpha(i, n, n_peek, alpha, rho)
        obrien_fleming_alpha.append(of_alpha)

    bonferroni_correction_alpha = [alpha/len(n_peek)] * len(n_peek)
    alpha_comparison = pd.DataFrame({'peek_window':np.arange(1,8), 'bonferroni_correction': bonferroni_correction_alpha, 'obrien_fleming': obrien_fleming_alpha})

    plt.figure(figsize=(10, 6))
    plt.plot(alpha_comparison['peek_window'], alpha_comparison['bonferroni_correction'], label='Bonferroni Correction', color='blue')
    plt.plot(alpha_comparison['peek_window'], alpha_comparison['obrien_fleming'], label="O'Brien-Fleming", color='red')
    plt.xlabel('Peek Window')
    plt.ylabel('Adjusted Alpha')
    plt.title('Comparison of Bonferroni Correction Alpha and O\'Brien-Fleming Alpha')
    plt.legend()

    plt.show()

    return alpha_comparison

alpha_comparison = compare_alpha(n, n_peek, alpha=0.05, rho=0.5)
display(alpha_comparison)
```

### Mixture Sequential Probability Ratio Test
- Simulation: use 100 runs

#### mSPRT - Simulation

```python
def calculate_tau(alpha, v_ctrl, v_trt, n_ctrl, n_trt):
    """
    To Compute the mixing parameter (tau) for the mSPRT, given:
    
    - alpha: Significance level (two-sided)
    - v_ctrl, v_trt: Variances of control & treatment groups
    - n_ctrl, n_trt: Sample sizes of control & treatment
    Returns:
    - revised_tau: The mixing parameter (tau)
    """
    z_alpha = sp.norm.ppf(1 - alpha / 2)   # Two-sided z critical
    revised_tau = (z_alpha ** 2) * ((v_ctrl / n_ctrl) + (v_trt / n_trt))
    return revised_tau


def mSPRT_statistic(X_ctrl, X_trt, tau, alpha):
    """
    To Compute the mSPRT decision, effect estimate (delta), and CI width.
    
    - X_ctrl, X_trt: control & treatment observations
    - tau: mixing parameter
    - alpha: significance level (two-sided)
    
    Returns:
    - decision: Boolean indicating if H0 is rejected
    - delta: ATE
    - ci: confidence interval
    """
    n_ctrl, n_trt = len(X_ctrl), len(X_trt)
    xbar_ctrl, xbar_trt = np.mean(X_ctrl), np.mean(X_trt)

    # Variance of the mean difference
    v_ctrl = np.var(X_ctrl, ddof=1)
    v_trt = np.var(X_trt, ddof=1)
    V = (v_ctrl / n_ctrl) + (v_trt / n_trt)

    # Mean difference
    delta = xbar_trt - xbar_ctrl

    # CI width components
    denominator = np.sqrt(V * (V + tau) / tau)
    threshold = np.sqrt(-2 * np.log(alpha) - np.log(V / (V + tau)))
    ci = denominator * threshold  # half-width of CI

    # Lower & upper bounds
    lower_bound = delta - ci
    upper_bound = delta + ci

    # Reject H0 if 0 is outside the CI
    decision = (lower_bound > 0) or (upper_bound < 0)

    return decision, delta, ci


def run_simulation(mu_trt, sigma_trt, n_ctrl, n_trt, mu_ctrl, sigma_ctrl, alpha, num_simulations):
    """
    To run a sequential mSPRT simulation for a given A/B test setup.
    
    - mu_trt, sigma_trt: True mean & std of the treatment group
    - mu_ctrl, sigma_ctrl: True mean & std of the control group
    - n_ctrl, n_trt: Sample sizes for control & treatment
    - alpha: Significance level
    - num_simulations: Number of simulation runs

    Returns:
    - df_results: A DataFrame of results with columns:
      ['simulation', 'step', 'decision', 'delta', 'ci', 'length']
    - check_performance_msprt(df_results): 
    """
    results = []
    false_positives = 0
    true_positives = 0

    for sim_id in range(num_simulations):
        # generate data for control & treatment
        X_ctrl, X_trt = generate_data(n_ctrl, n_trt, mu_ctrl, sigma_ctrl, mu_trt, sigma_trt)

        first_reject_h0 = False
        length = n_trt  # track earliest step that rejected H0

        # sequential decision at each step
        for i in range(2, min(n_ctrl, n_trt) + 1):
            # sub-sample up to step i
            X_ctrl_step = X_ctrl[:i]
            X_trt_step = X_trt[:i]

            # compute tau
            v_ctrl_step = np.var(X_ctrl_step, ddof=1)
            v_trt_step = np.var(X_trt_step, ddof=1)
            tau = calculate_tau(alpha, v_ctrl_step, v_trt_step, n_ctrl, n_trt)

            # compute decision, ATE and CI
            decision, delta, ci = mSPRT_statistic(X_ctrl_step, X_trt_step, tau, alpha)

            # update the earliest reject step
            length = np.minimum(length, i) * decision + length * (1 - decision)
            results.append({
                'simulation': sim_id,
                'step': i,
                'decision': decision,
                'delta': delta,
                'ci': ci,
                'length': length
            })

            # log the first time H0 is rejected
            if decision and not first_reject_h0:
                if mu_trt == mu_ctrl:
                    false_positives += 1
                else:
                    true_positives += 1
                first_reject_h0 = True

    # results
    df_results = pd.DataFrame(results)
    false_positive_rate = false_positives / num_simulations
    power = true_positives / num_simulations

    # average length across simulation trials
    avg_length = df_results[df_results['step'] == 1400]['length'].mean()

    if mu_trt == mu_ctrl:
        print(f"False Positive Rate (Reject H0 when true): {false_positive_rate:.4f}")
        print(f"Average length of conclusion: {avg_length:.0f}")
    else:
        print(f"Power (Reject H0 when false): {power:.4f}")
        print(f"Average length of conclusion: {avg_length:.0f}")

    return df_results, check_performance_msprt(df_results)



# transform df_results for `plot_power_fpr_msprt`.
def check_performance_msprt(df):
    """
    To transform the output of simulation.
    
    - df: The dataframe of simulation output

    Returns:
    - df_melted: A dataframe that can be used to plot the trend of Power and FPR
    """
    df = df.sort_values(['simulation', 'step'])
    df['decision_cummax'] = df.groupby('simulation')['decision'].cummax()
    df_pivoted = df.pivot(index='simulation', columns='step', values='decision_cummax').reset_index()
    df_pivoted["idx"] = 1
    df_pivoted = df_pivoted.groupby("idx", as_index=False).mean()

    df_melted = df_pivoted.melt(
        id_vars=['idx', 'simulation'],
        var_name='peek',
        value_name= 'result'
    ).drop(columns = 'simulation')

    return df_melted


# this is to plot one simulation run to visualize the CI.
def plot_mSPRT(df, ate, selected_simulation):
    """
    To demo how the CI is shrinking when the sample size increases.
    
    """
    df_filtered = df[(df['simulation'] == selected_simulation) & (df['step']>=100)]
    sns.set_style("white")

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(df_filtered['step'], df_filtered['delta'], label='Delta')

    ax.fill_between(df_filtered['step'], df_filtered['delta'] - df_filtered['ci']
                    , df_filtered['delta'] + df_filtered['ci'], color='gray', alpha=0.3)
    ax.axhline(0, color='red', linestyle='--')

    ax.set_ylabel('Average Treatment Effect')
    ax.set(xlim=(100, 1400), xlabel="Step", 
           title= f"Average Treatment Effect and Confidence Intervals (Actual ATE = {ate})")
    
# this is to plot the trend of Power and FPR as we do for other methods.  
def plot_power_fpr_msprt(df_perf_aa, df_perf_ab, method, alpha=0.05, beta=0.8):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 3))

    #plot power
    sns.lineplot(df_perf_ab, x="peek", y="result", c="C1", ax=ax1)
    ax1.axhline(y=beta, ls="--", lw=2, c="C1")
    ax1.set(title=f"Power - {method}", xlabel = '# Peekings', ylabel="Power", ylim=(0.00, 1.05))

    #plot false positive rate
    sns.lineplot(df_perf_aa, x="peek", y="result", c="C2", ax=ax2)
    ax2.axhline(y=alpha, ls="--", lw=2, c="C2")
    ax2.set(title=f"False Positive - {method}", xlabel = '# Peekings', 
                ylabel="False Positive Rate", ylim=(0.0, 0.5))
    
```

```python
# when H0 is true
n_ctrl = n_trt = 1400
mu_trt = mu_ctrl =0
sigma_ctrl = sigma_trt = 6.68
alpha = 0.05
num_simulations = 1000

# run functions
df_results_0, df_perf_0 = run_simulation(mu_trt, sigma_trt, n_ctrl, n_trt, mu_ctrl, 
                              sigma_ctrl, alpha, num_simulations)
plot_mSPRT(df_results_0, mu_trt-mu_ctrl, 50)
```

```python
# when H0 is not true
n_ctrl = n_trt = 1400
mu_ctrl, mu_trt = 0, 1
sigma_ctrl = sigma_trt = 6.68
alpha = 0.05
num_simulations = 1000

# run functions
df_results_1, df_perf_1 = run_simulation(mu_trt, sigma_trt, n_ctrl, n_trt, mu_ctrl, sigma_ctrl, alpha, num_simulations)
plot_mSPRT(df_results_1, mu_trt-mu_ctrl, 50)
```
